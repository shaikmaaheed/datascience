# Percentiles

To calculate percentile rank of given x value


percentile rank of x = (no.of values below x/n) * 100


find x value at given percentile = (percentile/100) * (n+1) --> element number in list

# Removing the outliers

We can use box plot to have a look on lower values and higher values

           Lower Fench   <------>   Higher Fench
           interquarle range = (Q3-Q1)


```python
x = [x for x in range(1,1000)]
x = [2000] + x + [-1500]
display(x)

import seaborn as sns

sns.boxplot(x)
```

Observe the IQR from around 0 - 1000, Q2 at 500 and outliers -1500 and 2000 are detected automatically

calculating lower fence and higher fence

Lower fence = Q1 - 1.5(IQR) ----> less than this fench is considered as outlier


Higher fence = Q3 + 1.5(IQR) ----> more than this fench is considered as outlier

```python
# calculating practically


import numpy as np


l1 = [40, 34, 54, 12, 55, 66, 77, 23, 56, 65, 89, 90, 100, 11, 33, 43, 77]

min, Q1, Median, Q3, max = np.quantile(l1, [0, .25, .5, .75, 1])

display(min, Q1, Median, Q3, max)
```

```python
# if any value crossed this limit then treated as outlier

IQR = Q3 - Q1

lower_fence = Q1 - 1.5*(IQR)
higher_fence = Q3 + 1.5*(IQR)

lower_fence, higher_fence
```
